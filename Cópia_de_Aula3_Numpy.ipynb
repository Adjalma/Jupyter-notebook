````python
# Importando a biblioteca

import numpy as np
````

````python
# Criando arrays

a = np.arange(1, 16)
print(a)

zeros = np.zeros_like(a)
print(zeros)
````

A operação `np.zeros_like` cria uma matriz com 0's com o mesmo tamanho da matriz de entrada (matriz `a`).

````python
# Multiplicando arrays

b = a * zeros
print(b)
````

````python
# Dividindo arrays

c = a / zeros
print(c)
````

````python
# Somando uma constante

d = a + 10
print(a)
print(d)
````

````python
# Multiplicando uma constante

e = a * 10
print(a)
print(e)
````

````python
# Produto matricial entre duas arrays

x = np.arange(1, 16)
x = x.reshape(3, 5)
print(x)

z = x * 10
print(z)

x.shape
````

DATA_ANALYSIS-Page-3.drawio.png

````python
w = z.transpose()
w.shape
````

DATA_ANALYSIS-Page-8.drawio.png

````python
array_produto_matricial = x.dot(w)
print(array_produto_matricial)

array_produto_matricial.shape
````

- **Observação**

Para fazer o produto matricial precisamos garantir as propriedades matemáticas *(número de colunas da primeira matriz = número de linhas da segunda matriz)*, por isso fizemos o transpose da segunda matriz.
