# Introdução

O que são projetos de ciencia de dados?

Projetos são conjuntos de ações em um espaço de tempo que tem como finalidade um objetivo ou meta definido no início do projeto. A definição de projeto segundo o PMBOK é “Segundo o PMBOK, um projeto é um esforço temporário empreendido para criar um produto, serviço ou resultado exclusivo. Ou seja, um projeto é tudo aquilo que precisamos realizar para gerar algo novo: seja uma casa, um sistema informatizado, um estudo/pesquisa, um trabalho de conclusão de curso, uma contratação ou um compra importante. Lembre-se que ser “temporário” significa que os projetos devem ter um início e um término definidos; não significa de curta duração.” 

Um projeto de ciência de dados nada mais é do que uma análise apurada de um conjunto de dados. Essas análises muitas vezes tem o intuíto de criar um modelo onde são feitas estimativas futuras para a tomada de decisões mais assertivas ou otimizar suas operações. Dentro desses projetos são utilizadas muitos tipos de tecnologias e técnicas, podemos citar as técnologias de machine learning, onde é possível ensinar a máquina a prever cenários e a propor a melhor tomada de decião. Mas para entender melhor algumas definições, técnicas e metodologias, vamos estudar alguns tópicos a seguir

# **Cultura Data Driven**

## ***O que é Cultura Data Driven?***

A tradução de Data Driven é ‘Orientado a dados’, isso quer dizer que todas as decisões dentro de uma empresa que possui uma cultura Data Driven é orientada com dados, ou seja, é necessário analisar os dados gerados por determinado processo para tomar decisões mais assertivas.

## ***Quais são as vatangens da utilização da Cultura Data Driven?***

- Decisões mais assertivas
- Análises direcionadas
- Maior confiabilidade no processo
- Transparência

## ***O que é Data mining?***

- Processo de extração e descoberta de padrões em grandes conjunto de dados, envolvendo métodos de Machine Learning, estatística e sistema de banco de dados.
- O Data Mining é basicamente a extração de valor dos dados e geração de padrões de comportamento

## ***O que é Machine Learning na ciência de dados?***

Machine Learning é o processo de ensinamento da máquina para detectar padrões nos dados. Com essa detecção podemos criar modelos em machine learning para prever possíveis cenários e assim se preparar para tomar as devidas decisões.

## ***O que é Ciência de dados?***

Ciência de dados é o campo que visa descobrir insight dos dados, utilizando estatística, Machine Learning e análise de dados.

# **CRISP-DM (Cross Industry Standard Process for Data Mining)**

## ***Por que precisamos de um framework para realizar projetos?***

Precisamos de uma metodologia para ajudar a seguir um passo a passo, pois assim conseguiremos padronizar todo o processo que realizamos e analisa-los separadamente, para verificar qual foi a linha de raciocínio até a chegada no resultado final.

## ***Quais metodologias existem para dados?***

KDD, SEMMA e CRISP-DM

## ***O que é o CRISP-DM?***

O CRISP-DM (Cross Industry Standard Process for Data Mining), criado em 1996, é um framework de projetos voltado para ciência de dados. Possui etapas para analise de dados e entendimento de negócio.

### ***Modeling***

É a etapa escolhe-se a melhor modelagem dos dados. Nessa etapa podemos separar em 4 principais tasks, que são:
- **Seleção da técnica de modelagem:** Etapa na qual se determina quais técnicas de modelagem serão utilizadas, por exemplo: regressão, classificação/recomendação. É preciso identificar também qual algoritmo será aplicado, por exemplo: regressão utilizando redes neurais com back propagation. É preciso documentar a técnica que será utilizada. Se for utilizar mais de uma técnica, documente cada uma delas. É interessante utilizar mais de uma técnica pois assim poderá comparar e selecionar o melhor modelo. Muitas técnicas de modelagem assumem premissas sobre os dados, como por exemplo, todos os atributos apresentam uma distribuição uniforme, ou nenhum valor nulo é aceito, etc. É preciso deixar claro quais são as premissas assumidas para cada técnica a ser utilizada.
- **Principais perguntas:** Quais são os tipos de variáveis que se está trabalhando? De acordo com o objetivo do projeto e com os dados disponíveis qual técnica faz mais sentido? Por quê? Quais premissas foram assumidas para cada técnica e por quê? Pensando no negócio, quais os critérios para selecionar a técnica?
- **Geração de design teste:** Antes de construir um modelo é preciso gerar um procedimento ou mecanismo para testar a qualidade e a validade do modelo gerado. Assim, para conseguir gerar o modelo, testá-lo e validá-lo é preciso separar os dados em 3 conjuntos: treino, teste e validação. Basicamente é preciso utilizar uma parte dos dados para treinar o modelo, outro conjunto para testá-lo e por fim um conjunto para validar. Por exemplo: em modelos supervisionados como classificação é comum utilizar frequência de erros como medida de qualidade. Para isso costuma-se separar o dataset em treino e teste. O modelo é construído utilizado o dataset de treino e a qualidade é estimada utilizando o dataset de teste. Como o modelo não teve contato com o dataset de teste, a assertividade que ele terá com o teste será uma medida válida da qualidade do modelo.
- **Principais perguntas**: Como o modelo será treinado, testado e avaliado? O dataset será dividido em quantos conjuntos? Por quê? Como o dataset será divido?
- **Modelo de construção:** Rode a ferramenta de modelagem no dataset preparado para construir o modelo. Por mais glamoroso que pareça, isso pode ser apenas a execução de algumas linhas de código como “reg = LinearRegression (). Fit (X, y)”. Quando se constrói modelos, as ferramentas fornecem opções de configuração dos parâmetros que podem ser ajustados. Esses ajustes terão impacto direto na estrutura do modelo. Liste os parâmetros utilizados bem como os valores definidos para cada um deles, deixando claro o raciocínio utilizado nessas escolhas. - **Avaliação do modelo**: Nesta etapa se avalia os modelos gerados tanto de um ponto de vista técnico (ex: acurácia) quanto do ponto de vista do negócio (levando em consideração os objetivos de negócio e seus critérios de sucesso). Geralmente, vários modelos competem entre sí e o cientista de dados precisa interpretar os resultados do modelo com base no conhecimento do domínio, nos critérios de sucesso pré-definidos e no design do teste. Faça um resumo dos resultados, listando a qualidade dos modelos gerados e ranqueie a qualidade de cada modelo com relação aos outros. Nessa etapa, provavelmente será necessário revisar os parâmetros adotados na construção do modelo, fazer novos ajustes e rodar novamente a construção dos modelos. Embora o guia CRISP-DM sugira "iterar a construção e avaliação do modelo até que você acredite fortemente que encontrou o(s) melhor(es) modelo(s)", na prática, as equipes devem continuar iterando até encontrar um modelo "bom o suficiente", prossiga com o CRISP - Ciclo de vida do DM e, em seguida, melhore ainda mais o modelo em iterações futuras.

- **Principais perguntas**: Na avaliação do grupo, o que deu certo e o que deu errado?O que foi bem feito? Os usuários finais estão satisfeitos? Precisam de suporte adicional? O que precisa ser melhorado?

## ***Quais as vantagens da utilização do CRISP-DM?***


- Maior controle do projeto
- Melhor definição de etapas
- Maior transparência e visibilidade das análises
- Movimentação cíclica

# **KDD (Konowledge Discovery in Database)**

## ***O que é o KDD?***

É um framework para trabalhar com o data mining. Essa metodologia deu origem ao CRISP-DM, esse framework trabalha muito mais o tratamento do dado, enquanto o CRISP-DM trabalha mais a área de negócio.

Untitled(4).png

## ***Quais são as etapas do KDD?***

- **Selection**: Nessa etapa é feita a coleta de dados para o projeto.
- **Processing**: Nessa etapa realiza-se a analise da qualidade dos dados, limpeza e remoção de dados inconsistentes.
- **Transformation**: Nessa etapa são selecionadas técnicas de transformação, como por exemplo: normalização, agregação ou redução de dados.
- **Data Mining**: Nessa etapa é realizado a construção de modelo (ML ou estátistico) aplicando a técnica para extração de valor dos dados.
- **Interpretation/Evaluation: Nessa etapa avalia-se a** saúde do modelo e os resultados obtidos

# **SEMMA**

## ***O que é o SEMMA?***

Criado pelo SAS Enterprise Miner e tem como foco a criação dos modelos para mineração de dados. Ele não abrange muito a fase de negócio.

Untitled(5).png

## ***Quais são as etapas do SEMMA?***

- **Amostra**: Nessa etapa realiza-se a coleta e separação de uma amostra dos dados que será avaliada.
- **Explorar**: Nessa etapa realiza-se análises de estatísticas e avaliar normalias que existem nesses dados.
- **Modificar**: Nessa etapa realiza-se a transformação dos dados e preparação para a aplicação do modelo.
- **Modelar**: Nessa etapa realiza-se a aplicção de técnicas de mineração de dados como Machine Learning, estatística, etc.
- **Avaliar**: Nessa etapa realiza-se a avaliação dos resultados obtidos e realização de alterações, caso necessário.

# **ASUMA-DM (Analytics Solutions Unified Method for Data Mining)**

## ***O que é o ASUM-DM?***

Método lançado pela IBM em 2015 para suprir as deficiências do CRISP-DM e criar uma integração maior com a metodologia agil.
Untitled(6).png

## ***Quais as etapas?***

- **Analyse**: Nessa etapa verifica-se quais predições preciso fazer, quais problemas preciso resolver e quais variaveis preciso classificar. Defini-se também os recursos não funcionais como performace, velocidade e usabilidade do modelo.
- **Design:** Define todos os componentes que fazem parte dessa solução, além dos modelos e banco de dados.
- **Configure & Build:** Nessa etapa realiza-se configuração, construção e integração dos componentes da solução.
- **Operate & Optimize:** Nessa etapa realiza-se a manutenção e monitoramento do modelo.

# **MLCanvas**

## ***O que é o Machine Learning Canvas?***

Ferramenta para a facilitação das etapas para a construção de modelo. Não substitui os frameworks de ciência de dados.
Untitled(7).png

## ***Como preencher o Machine Learning Canvas?***

- **Divisão**: O MLCanvas é dividido em 4 grandes tópicos, que são: O que, Por que, Quem e Como?
- **Centro do MLCanvas:** No centro responderemos sucintamente as perguntas O que, Por que, Quem
- **Como aprender (How to learn)**: Vão estar as informações de dados e treinamentos do projeto
    - **Data Sources:** Onde estão e quais bases vamos utilizar
    - **Collecting Data**: Como vamos coletar esses dados
    - **Features:** Quais características dos dados são importantes para o nosso modelo?
    - **Building Models:** Vamos especificar quando vamos construir e qual frequência que vamos avaliar e dar manutenção ao modelo
- **Como predizer (How to predict)**: Decisões e métricas que vamos avaliar
    - **ML Task:** O que temos que prever e classificar
    - **Decisions:** Como a área de negócios usará a predição para tomar uma decisão
    - **Offline Evaluation:** Definição de métrica de negócio para avaliar se o modelo está funcionando corretamente ou não
    - **Making Predictions:** Como o algoritmo vai funcionar? Frequência de atualização da predição
- **How well Evaluation:** Quão bem o modelo está funcionando?
    - **Live Evaluation and Monitoring**: Definir como monitorar o comportamento do modelo em produção
